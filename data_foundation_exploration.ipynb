```python
import pandas as pd 
```

```python
def load_data(file_path):
    df=pd.read_csv(file_path)
    return df
```

```python
def check_missing_values(df):
    return df.isnull().sum()
```

```python
def check_for_empty_strings(df):
    return (df['Text'].str.strip()=='').sum()
```

```python
def get_language_distribution(df):
    return df['Language'].value_counts()
```

```python
if __name__ == "__main__":
    DATA_FILE_PATH = r"C:\Users\nyasa\Desktop\ML PROJECTS\language-detection-system\Language Detection.csv"
    data = load_data(DATA_FILE_PATH)

    if data is not None:
        #print("--- First 10 Rows ---")
        #print(get_data_head(data, n=10))

        print("\n--- Missing Values ---")
        print(check_missing_values(data))

        print("\n--- Empty Strings ---")
        print(check_for_empty_strings(data))

        print("\n--- Language Distribution ---")
        print(get_language_distribution(data).head(10))
```
